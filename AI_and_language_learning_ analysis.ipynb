# Select the feature for next graph "What AI-tools are used to learn languages?" and cleaning data

```python
tools_series = df['–ö–∞–∫–∏–µ –ò–ò-–∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç—ã –≤—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∏ –¥–ª—è –∏–∑—É—á–µ–Ω–∏—è —è–∑—ã–∫–æ–≤? (–ú–æ–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ)  '].str.split(';').explode()
tools_series = tools_series.reset_index(drop=True)
tools_series = tools_series.drop(41)
tools_series = tools_series.drop(22)
tools_series.loc[23] = "Don't use"
tools_counts = tools_series.value_counts()
tools_counts = tools_counts.astype(int)
```

# Plot the graph

```python
custom_colors = ["#FF6B6B", "#1E90FF", "#FFD700", "#32CD32", "#8A2BE2", "#FF4500"]
plt.figure(figsize=(10, 5))
ax = sns.barplot(x=tools_counts.index, y=tools_counts.values, palette=custom_colors)
ax.set_yticks(np.arange(0, tools_counts.max() + 1, 3))
plt.xticks(rotation=45)
plt.xlabel("AI-tools")
plt.ylabel("Number of users")
plt.title("What AI-tools are used to learn languages?")
plt.savefig("ai_tools_usage.png", dpi=300, bbox_inches="tight")
plt.show()
```

# ChatGPT is the most popular AI language learning tool, followed by Duolingo and Grammarly. Other tools are used much less frequently, and some respondents do not use AI at all to learn languages.
